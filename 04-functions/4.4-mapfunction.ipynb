### The map() Function in Python


The map() function applies a given function to all items in an input list (or any other iterable) and returns a map 
object (an iterator). This is particularly useful for transforming data in a list comprehensively.

```python
def square(x):
    return x ** 2   
square(2)
```

```python
numbers = [1, 2, 3, 4, 5]
list(map(square, numbers))
```

```python
## Lamda function with map
numbers = [1, 2, 3, 4, 5]
list(map(lambda x: x ** 2, numbers))
```

```python
## Map multiple iterables
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

list(map(lambda x, y: x + y, numbers1, numbers2))
```

```python
## map to convert a list of strings to integers 
# Use map to convert strings to integers
str_numbers = ['1', '2', '3', '4', '5']
int_numbers = list(map(int, str_numbers))

print(int_numbers)  # Output: [1, 2, 3, 4, 5]
```

```python
words = ['hello', 'world', 'python']
upper_word = list(map(lambda x: x.upper(), words))
print(upper_word)  # Output: ['HELLO', 'WORLD', 'PYTHON']
```

```python
def get_name(person):
    return person['name']

people =[
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Charlie', 'age': 35}
]
list(map(get_name,people))
```

### Conclusion

The map() function is a powerful tool for applying transformations to iterable data structures. It can be used 
with regular functions, lambda functions, and even multiple iterables, providing a versatile approach to data
processing in Python. By understanding and utilizing map(), you can write more efficiency and readable code
