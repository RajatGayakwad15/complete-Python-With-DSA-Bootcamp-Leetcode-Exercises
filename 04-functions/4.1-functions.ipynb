```python
## why function ?
num = 24
if num%2 ==0:
    print("Even")
else:   
    print("Odd")
```

```python
def even_or_odd(num):
    if num%2 ==0:
        return "Even"
    else:
        return "odd"
    

even_or_odd(25)
```

```python
## Call this function
even_or_odd(24)
```

```python
## function with multiple parametes
def add(a,b):
    return a+b
result = add(2,3)
print(result)
```

```python
## Default parameters

def greet(name):
    print(f"hello {name}")

greet("Alice")
```

```python
def greet(name="Guest"):
    print(f"hello {name}")

greet()
```

```python
### Variable Length Arguments
### Positional and keywords arguments

def print_numbers(*args):
    for number in args:
        print(number)
print_numbers(1, 2, 3, 4, 5, "Rajat")
```

```python
### Keyword arguments
def print_details(**kargs):
    for key,value in kargs.items():
        print(f"{key } : {value } ")

print_details(name = "Rajat", age=30 , city="Kolhapur")        
```

```python
def print_details(*args, **kargs):
    for val in args:
        print(val)
    
    for key,value in kargs.items():
        print(f"{key} : {value}")

print_details("Hello", "World", name = "Rajat", age=30 , city="Kolhapur")
```
