## Lambda Functions in Python

Lambda functions are small anonymous functions defined using the lamda keyword. They can have any 
number of arguments but only one expressions. They are commonly used for short operations or as arguments
to higher-orders functions.

```python
#Syntax
lambda arguments: expression
```

```python
def add(x, y):
    return x + y
print(add(2, 3))  # Output: 5
```

```python
addition = lambda x, y: x + y
type(addition)  # Output: <class 'function'>
print(addition(2, 3))  # Output: 5
```

```python
def even(num):
      if num % 2 == 0:
        return True
    
even(4)
```

```python
even1 = lambda num:num%2 ==0 
print(even1(4))
```

```python
addition1 = lambda a, b, c: a + b + c
print(addition1(1, 2, 3))  # Output: 6
```

```python
## map() - it applies a function to all items in a list
numbers = [1, 2, 3, 4, 5]
def square(x):
    return x ** 2
square(2)
```

```python
list(map(lambda x: x ** 2, numbers))
```
