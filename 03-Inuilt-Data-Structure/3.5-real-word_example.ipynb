###Real-World Examples Using Lists in Python

 Lists are one of the most commonly used data structures in python, thanks to their versatility and ease of use.
 Here are several practicle example that illustrate their use in real-world scenarios

##Example 1. Manage A To Do List

create a To do List To keep Track OF tasks

```python
to_do_list=["Buy groceries", "Call mom", "Finish project", "Go for a run"]
## Adding a new task to the list
to_do_list.append("Read a book")
to_do_list.append("Clean the house")

## Removing a task from the list
to_do_list.remove("Go for a run")

## Checking if a task is in the list
if "Call mom" in to_do_list:
    print("Don't forget to call mom!")
print("to do list remaining")
for task in to_do_list:
    print(f"-{task}")
```

##Example 2: Organizing Student Grades
 
 

 create a list to store and calculate average grades for students

```python
## organizing student grades
grades=[85,92,78, 80, 90]

# Adding a new grades
grades.append(95)

# calculating the average grade
average_grade = sum(grades) / len(grades)

# Find the highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"Average grade: {average_grade}")
print(f"Highest grade: {highest_grade}")
print(f"Lowest grade: {lowest_grade}")
```

##Managing an Inventory

Use a list to manage inventory items ina store

```python
# Managing an inventory
inventory = ["apples", "oranges", "bananas", "grapes"]

# Adding new items to the inventory
inventory.append("pears")
inventory.append("watermelons")

# Removing an item from the inventory
inventory.remove("oranges")

# Checking if an item is in the inventory
if "bananas" in inventory:
    print("We have bananas in stock!")
else:
    print("We need to restock bananas.")

# Peinting the current inventory
print("Current inventory:")
for item in inventory:
    print(f"-{item}")
```

##Example 4: Collection User Feedback


Use a list to collect and analyze user feedback

```python
#Collecting user feedback
feedback = ["Great", "Could be better", "Excellent", "Not satisfied"]

# Adding new feedback
feedback.append("Average")

# Counting specific feedback
positive_feedback_count = sum(1 for comment in feedback if"great" in comment.lower() or "excellent" in comment.lower())
print(f"Poitive feedback count:{positive_feedback_count}")

# Priniting all feedback
for comment in feedback:
    print(f"-{comment}")
```
