### Sets

Sets are a builts-in data type in python used to store collections of unique items. They are unorderd, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests,eliminting duplicate entries, and performing matematical set operational like union, intersection,differenece , and symmetric difference .



```python
## Create a set 
my_set = {1, 2, 3, 4, 5}
print(my_set) # Output: {1, 2, 3, 4, 5}
print(type(my_set)) # Output: <class 'set'>
```

```python
my_set = set([1, 2, 3, 4, 5])
print(my_set) # Output: {1, 2, 3, 4, 5}
print(type(my_set)) # Output: <class 'set'>
```

```python
my_set = set([1, 2, 3, 4, 5, 6,1,2 ,3])
print(my_set) # Output: {1, 2, 3, 4, 5}
```

##Basic sets operation

```python
## Adding elements to a set
my_set.add(7)
print(my_set) # Output: {1, 2, 3, 4, 5, 6, 7} 
my_set.add(7)
print(my_set) # Output: {1, 2, 3, 4, 5, 6, 7}  
```

```python
## Removing elements from  a set
my_set = set([1, 2, 3, 4, 5])
print(my_set) # Output: {1, 2, 3, 4, 5}
my_set.remove(3)
print(my_set) # Output: {1, 2, 4, 5}
```

```python
my_set.remove(10) # This will raise a KeyError since 10 is not in the set
```

```python
my_set.discard(10) # This will not raise an error even if 10 is not in the set
print(my_set) # Output: {1, 2, 4, 5}
```

```python
## Pop method
removed_element = my_set.pop()
print(f"Removed element: {removed_element}") # Output: Removed element: 1 (or any other element from the set)
print(my_set) # Output: {2, 4, 5} (or the remaining elements in the set after popping)
```

```python
## Clear method
my_set.clear()
print(my_set) # Output: set()
```

```python
## Set Membership test
my_set = set([1, 2, 3, 4, 5])
print(3 in my_set) # Output: True
print(10 in my_set) # Output: False
```

```python
## Mathematical set operationsons
A = set([1,2,3,4,5])
B= set([4,5,6,7,8])

### union
union_set=A.union(B)
print(union_set) # Output: {1, 2, 3, 4, 5, 6, 7, 8}

### intersection
interaction_set=A.intersection(B)
print(interaction_set) # Output: {4, 5}

A.intersection_update(B)
print(A) # Output: {4, 5}
```

```python
A = set([1,2,3,4,5])
B= set([4,5,6,7,8])

### difference
difference_set=A.difference(B)
print(difference_set) # Output: {1, 2, 3}
```

```python
print(B.difference(A)) # Output: {6, 7, 8}   
```

```python
## Symmetric difference
sym_diff=A.symmetric_difference(B)
print(sym_diff) # Output: {1, 2, 3, 6, 7, 8}
```

```python
### Sets methods
set1={1,2,3 , 4,5}
set2={3,4,5}

## is subset 
print(set1.issubset(set2)) # Output: False
print(set2.issubset(set1)) # Output: True

## is superset]
print(set1.issuperset(set2)) # Output: True
print(set2.issuperset(set1)) # Output: False
```

```python
lst=[1,1,2,3,4,4,3]
print(lst) # Output: [1, 1, 2, 3, 4, 4, 3]
s1=set(lst)
print(s1) # Output: {1, 2, 3, 4}
```

```python
### counting unique word in text
text = "hello world hello"
words=text.split()

### convert list of words to set to get unique words
unique_words=set(words)
print(unique_words) # Output: {'hello', 'world'}
print(len(unique_words)) # Output: 2
```
