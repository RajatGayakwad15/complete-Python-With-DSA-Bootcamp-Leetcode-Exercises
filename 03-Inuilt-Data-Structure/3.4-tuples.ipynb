##Indroduction to Tuples
  
  Tuples are ordered collections of items that are immutable. They are similar to lists, but their immuatability makes them different .

```python
## creating a tuple
empty_tuple = ()
print(empty_tuple)  # Output: ()
print(type(empty_tuple))  # Output: <class 'tuple'>
single_element_tuple = (42,)
print(single_element_tuple)  # Output: (42,)
```

```python
lst=list()
print(type(lst))  # Output: <class 'list'>
tpl=tuple()
print(type(tpl))  # Output: <class 'tuple'>
```

```python
numbers = tuple([1, 2, 3, 4, 5])        
print(numbers)  # Output: (1, 2, 3, 4,5)
```

```python
list((1, 2, 3))  # Output: [1, 2, 3]
```

```python
mixed_tuple = (1, "hello", 3.14, [1, 2, 3], (4, 5))
print(mixed_tuple)  # Output: (1, 'hello', 3.14, [1, 2, 3], (4, 5))
```

```python
numbers
```

```python
print(numbers[0])  # Output: 1
print(numbers[1:4])  # Output: (2, 3, 4
print(len(numbers))  # Output: 5
print(3 in numbers)  # Output: True
print(numbers + (6, 7))  # Output: (1, 2, 3, 4, 5, 6, 7)
print(numbers[-1])  # Output: 5
print(numbers[::-1])  # Output: (5, 4, 3, 2, 1)
```

```python
## Tuple operations
numbers + mixed_tuple  # Output: (1, 2, 3, 4, 5, 1, 'hello', 3.14, [1, 2, 3], (4, 5))
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```

```python
## Immutable Nature of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned.
lst=[1, 2, 3 , 4]
print(lst)  # Output: [1, 2, 3, 4]
lst[0] = 10
print(lst)  # Output: [10, 2, 3, 4]
```

```python
numbers[1] = 20  # This will raise a TypeError: 'tuple' object does not support item assignment
```

```python
## Tuple methods
numbers.count(3)  # Output: 1
print(numbers.index(4))  # Output: 3
```

```python
## Packing and Unpacking tuples
## Packing
packed_tuple = 1, "hello", 3.14
print(packed_tuple)  # Output: (1, 'hello', 3.14
```

```python
## Unpacking a tuple 
a, b, c = packed_tuple
print(a)  # Output: 1
print(b)  # Output: 'hello'     
print(c)  # Output: 3.14
```

```python
## Unpacking with * operator
numbers = (1, 2, 3, 4, 5)
first,*middle,last =numbers
print(first)
print(middle)
print(last)
```

```python
numbers = (1, 2, 3, 4, 5)
*first, middle, last = numbers
print(first)  # Output: [1, 2, 3]
print(middle)  # Output: 4
print(last)  # Output: 5
```

```python
numbers = (1, 2, 3, 4, 5)
first, middle, *last = numbers
print(first)  # Output: 1
print(middle)  # Output: 2
print(last)  # Output: [3, 4, 5]
```

```python
## Nested Tuples
## Nested Lists
lst=[[1, 2], [3, 4], (5, "hello", 3.14)]
lst[2][0:2]
```

```python
nested_tuple = (1, (2, 3), (4, (5, 6)))
print(nested_tuple)  # Output: (1, (2, 3), (4, (5, 6)))
print(nested_tuple[1])  # Output: (2, 3)
print(nested_tuple[2][1][0])  # Output: 5
```

```python
## iterating over nested tuples
nested_tuple = (("a","b"), ("c","d"))
for sub_item in nested_tuple:
    for item in sub_item:
        print(item,end="")
    print()
```
