### Introduction to Dictionaries
 Dictionaries are unorded collections of item. They store data in key-value pairs. Keys must br unique and
  immutable (e.g., strings, numbers, or tuples), while values can be of any type.


```python
### Creating Dictinaries 
empty_dict = {}
print(empty_dict)   
# Output: {}
print(type(empty_dict))
```

```python
empty_dict = dict()
print(empty_dict)   
# Output: {}
```

```python
student ={
    "name": "rajat",
    "age":23,
    "grade":"A"

}
print(student)
```

```python
#Single key alwasy used

student ={
    "name": "rajat",
    "age":23,
    "grade":"A",
    "name": "tanmay"

}
print(student)
```

```python
## Accessing Values in a Dictionary
print(student['grade'])
print(student['name'])


## Accessing using get() method 
print(student.get('grade'))
print(student.get('name'))
print(student.get('address'))  # Output: None
print(student.get('address', 'Not Found'))  # Output: Not Found
```

```python
## Modifying Dictionary Values
student['grade'] = 'B' 
print(student['grade'])  # Output: B
student['age'] += 1
print(student['age'])  # Output: 24
student['address'] = '123 Main St'
```

```python
## Deleting Key-Value Pairs
del student['grade']
print(student)
```

```python
## Dictinoary Methods
print(student.keys())  # Output: dict_keys(['name', 'age', 'address'])
print(student.values())  # Output: dict_values(['tanmay', 24, '123 Main St'])
print(student.items())  # Output: dict_items([('name', 'tanmay'), ('age', 24), ('address', '123 Main St')])
```

```python
##Shallow copy of a dictionary
student_copy = student.copy() # or student_copy = student
print(student_copy)  # Output: {'name': 'tanmay', 'age': 24, 'address': '123 Main St'}  
```

```python
student_copy['name'] = 'rajat'
print(student_copy)  # Output: {'name': 'rajat', 'age': 24, 'address': '123 Main St'}
print(student)  # Output: {'name': 'tanmay', 'age': 24, 'address': '123 Main St'}
```

```python
### Iterating Over Dictionaries
## You can use loops to iterate over dictinaries , keys, values or items
for key in student.keys():
    print(key)
```

```python
## iterate over values 
for value in student.values():
    print(value)
```

```python
## Iterate over key value pairs 
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
## Nested Dictionaries
students = {
    "student1": {
        "name": "tanmay",
        "age": 24
    },
    "student2": {
        "name": "rajat",
        "age": 23
    }
}
```

```python
##   Accessing nested dictionary values
print(students['student1']['name'])  # Output: tanmay   
print(students['student2']['age'])   # Output: 23
```

```python
## Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}:")
    for key, value in student_info.items():
        print(f"  {key}: {value}")
```

```python
## Dictionary Comprehension
sqaures ={ x:x**2 for x in range(5)}
print(sqaures)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```

```python
## conditional dictionary comprehension
even ={x: x**2 for x in range(5) if x % 2 == 0}
print(even)  # Output: {0: 0, 2: 4, 4: 16}
```

```python
## Practical Example: 
## Use a dictionary to count the frequency of elements in a list
numbers = [1, 2, 2, 3, 3, 3, 4]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)  # Output: {1: 1, 2: 2, 3: 3, 4: 1}
```

```python
## Merge Dictionaries
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
merged = {**dict1, **dict2}
print(merged)  # Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}
```
