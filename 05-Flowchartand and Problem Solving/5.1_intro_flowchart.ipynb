### Flowchart

1) A flowchart is a graphical representaion of a process, system or algorith.
2) It uses symbols and arrows to show the sequence of steps needed to complete a task.
3) Flow charts are useful in visualising how to solve problem and make the process easier to flow

#### Common Flowchart Symbols

- **Oval (Terminator)** → Start / End of the process  
- **Rectangle (Process)** → A step or instruction  
- **Diamond (Decision)** → A condition (Yes/No, True/False)  
- **Arrow (Flowline)** → Direction of flow  

#### Additional Flowchart Symbols

- **Parallelogram (Input/Output)** → Used for taking input or displaying output  
- **Circle (Connector)** → Connects different parts of the flowchart (on the same page)  
- **Off-page Connector** → Links flowchart parts across different pages  
- **Double Rectangle (Predefined Process)** → A function or subroutine  
- **Document Symbol** → Represents a document or report  
- **Cylinder (Database/Storage)** → Data stored in a database  
- **Trapezoid (Manual Input)** → User enters data manually  
- **Delay Symbol** → Represents waiting or delay in process  

### Conclusion

Flowcharts make complex processes simple and clear by visually representing each step, making them an important tool in programming and system design.
