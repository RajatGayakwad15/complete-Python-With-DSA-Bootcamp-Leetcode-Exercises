```python
## Whilke loop
## The while loop continues to exexute as long as the condition is true

count=0
while count<5:
    print(count)
    count=count+1
```

```python
count=0
while count%2==0:
    print(count)
    count=count+1
```

```python
## Loop control statements
## break : used to exit the loop

for i in range(1,10):
    if i==5:
        break
    print(i)
```

```python
## continue : used to skip the current iteration and move to the next iteration
for i in range(1,10):
    if i==5:
        continue
    print(i)
```

```python
## pass : used when we want to write an empty loop or function
for i in range(1,10): 
    if i==5:
        print("5 is skipped")
        pass
    print(i)
```

```python
## Nested loops : loop inside another loop
for i in range(1,4):
    for j in range(1,4):
        # print(i,j)
        print(f"i:{i} and  j:{j}")
```

```python
## Example - Calculate the sum of first n natural numbers using a while loop

n=int(input("Enter a number:"))
sum=0
i=1
while i<=n:
    sum=sum+i
    i=i+1
print(f"The sum of first {n} natural numbers is {sum}")
```

```python
## Example - Calculate the sum of first n natural numbers using a for loop

n=int(input("enter any number"))
sum=0;
j=1
for i in range (j,n+1):
    sum+=i
print(f"The sum of first {n} natural numbers is {sum}") 
```

```python
## Example - Prime numbers between 1 to 100

for num in range(1,101):
    if(num>1):
        for i in range (2,num):
            if(num%i==0):
                break
        else:
            print(num)
            

```
