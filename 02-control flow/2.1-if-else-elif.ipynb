```python
## elif
## The elif statement allows you to check multiple conditions. It stands for "else if"

age=20

if age<13:
  print("You are child")
elif age<18:
  print("You are Teenager")
else:
  print("You are an adult")    
 
```

```python
## Nested Conditional Statement
## You can place one or more if , elif, or else statement insdie another if, elif, or else statemnt to create nested conditional statments.

## number even , odd, negative

num=int(input("Enter a number: "))

if num>0:
  if num%2==0:
    print("The number is even")
  else:
    print("The number is odd")
else:
  print("The number is zero and negative")
```

```python
## Pratical Example

## Determine if  a year is a leap year using nested condition statemnt

year=int(input("Enter a year: "))
if year%4==0:
  if year%100==0:
    if year%400==0:
      print("The year is a leap year")
    else:
      print("The year is not a leap year")
  else:
    print("The year is a leap year")   
```

```python
## Assignment
## Simple Calculater
## Take user Input

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")
if operator=="+":
  result=num1+num2
  print("The result is: ", result)
elif operator=="-":
  result=num1-num2
  print("The result is: ", result)
elif operator=="*":
  result=num1*num2
  print("The result is: ", result)
elif operator=="/":
  if num2!=0:
    result=num1/num2
    print("The result is: ", result)
  else:
    print("Error: Division by zero is not allowed")
else:
  print("Invalid operator")
                       
```
